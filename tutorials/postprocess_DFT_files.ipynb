# VASP files

```python
base_path = 'VASP_io_files/'

# this is if you ran your VASP calculations in folders called "pol_orig/" and "np_trans/"
pol_POSCAR_file = base_path+'pol_orig/POSCAR'
np_POSCAR_file = base_path+'np_trans/POSCAR'
pol_wavecar = base_path+'pol_orig/WAVECAR'
np_wavecar = base_path+'np_trans/WAVECAR'
potcar_file = base_path+'np_trans/POTCAR'
pol_OUTCAR = base_path+'pol_orig/OUTCAR'
np_OUTCAR = base_path+'np_trans/OUTCAR'    
pol_directory = base_path+'pol_orig/'
np_directory = base_path+'np_trans/'
    

vasp_parse_dict = bfd.VASPParser.vasp_parser(pol_POSCAR_file,
                                             np_POSCAR_file,
                                             pol_wavecar,
                                             np_wavecar,
                                             potcar_file,
                                             pol_directory,
                                             np_directory)

vasp_ovlps = bfd.Overlaps.Overlaps(vasp_parse_dict)
vasp_pol, vasp_svd_dict = vasp_ovlps.compute_polarization() # formal polarization between two reference structures in uC/cm^2
```

# Quantum ESPRESSO files

```python
base_path = 'QE_io_files/'
np_xml_file = base_path+'np_trans/calculations_555/BaTiO3_np_trans.xml' # need .xml files from each structure
pol_xml_file = base_path+'pol_orig/calculations_555/BaTiO3_pol_orig.xml' # need .xml file from each structure
np_wfcn_path = base_path+'np_trans/calculations_555/BaTiO3_np_trans.save/' # need folder wavefunctions saved in .hdf5 format
pol_wfcn_path = base_path+'pol_orig/calculations_555/BaTiO3_pol_orig.save/' # need folder wavefunctions saved in .hdf5 format
np_out_file = base_path+'np_trans/PbTiO3_np_trans_k555_scf_nosym.out' # need .out file (from either structure)

qe_parse_dict = bfd.QEParser.qe_parser(pol_xml_file,
                    np_xml_file,
                      pol_wfcn_path,
                      np_wfcn_path,
                      np_out_file)

qe_ovlps = bfd.Overlaps.Overlaps(qe_parse_dict)
qe_pol, qe_svd_dict = qe_ovlps.compute_polarization()
```
