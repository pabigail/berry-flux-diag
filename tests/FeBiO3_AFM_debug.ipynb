```python
base_path_qe = '../../new_bfd_code/qe_io/FeBiO3/FeBiO3_AFM/dft_u_calcs/scf_no_symm/'
np_xml_file = base_path_qe+'np_trans_1/calculations_k555_u_0/febio3_np_trans_1_scf.xml'
pol_xml_file = base_path_qe+'pol_orig/calculations_k555_u_0/febio3_pol_orig.xml'
np_wfcn_path = base_path_qe+'np_trans_1/calculations_k555_u_0/febio3_np_trans_1_scf.save/'
pol_wfcn_path = base_path_qe+'pol_orig/calculations_k555_u_0/febio3_pol_orig.save/'
np_out_file = base_path_qe+'np_trans_1/FeBiO3_AFM_k555_np_trans_1_scf_u_0.out'
```

```python
qe_dict = bfd.QEParser.qe_parser(pol_xml_file,
                        np_xml_file,
                        pol_wfcn_path,
                        np_wfcn_path,
                        np_out_file)
```

```python
Overlaps = bfd.Overlaps.Overlaps(qe_dict)
```

```python
pol, debug_dict = Overlaps.compute_polarization()
```

```python
lattice = Overlaps.pol_struct.lattice
fcoords_pol = Overlaps.pol_struct.frac_coords
fcoords_np = Overlaps.np_struct.frac_coords
pbc_shortest_vecs = bfd.Overlaps.pbc_shortest_vectors(lattice, fcoords_np, fcoords_pol)
cart_coords_for_ion_calc = []
```

```python
for i in range(len(pbc_shortest_vecs)):
    cart_coords_for_ion_calc.append(pbc_shortest_vecs[i][i])
frac_coords_for_ion_calc = lattice.get_fractional_coords(cart_coords_for_ion_calc) 
```

```python
tot_ionic = []
for site, frac_coord in zip(Overlaps.pol_struct, frac_coords_for_ion_calc):
    zval = Overlaps.zval_dict[bfd.utils.extract_letters(str(site.specie))]
    tot_ionic.append(Overlaps.calc_ionic(frac_coord, Overlaps.pol_struct, zval))
ion_diff =  np.sum(tot_ionic, axis=0)
```
